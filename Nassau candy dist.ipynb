## Cost Time Trade-offs by Ship Mode (description) 

Based on the aggregated data, we can observe the following trade-offs between shipping lead time and profitability (gross profit and sales):

1. Standard Class

Usage: Most popular, with 2445 orders.

Speed: Fastest average delivery time (1354.83 days).

Profitability: Not the highest profit per order, but because it’s used so much, it adds up to a big overall profit. Cheap and reliable, which makes it the main choice.

2. Second Class

Usage: Medium usage, with 799 orders.

Speed: Slightly slower than Standard (1372.39 days).

Profitability: Best performer—highest average Sales (14.03) and Gross Profit (9.24) per order. Customers seem willing to pay more for this option.

3. First Class

Usage: Less used, with 639 orders.

Speed: Slower than Standard and Second Class (1385.60 days).

Profitability: Lower average Sales (13.65) and Gross Profit (8.99) compared to Second Class. Surprising, because “First Class” usually means faster and more expensive, but here it’s worse.

4. Same Day

Usage: Least used, with only 247 orders.

Speed: Slowest average delivery (1405.35 days). This doesn’t match the name “Same Day,” which suggests a data problem or mislabeling.

Profitability: Lowest average Sales (12.65) and Gross Profit (8.35). Weakest performer overall.

🔑 Key Observations
Names don’t match performance: “First Class” and “Same Day” are actually slower and less profitable, which is the opposite of what their names suggest.

Standard Class dominates: It’s the fastest and most used, making it the backbone of operations.

Second Class is most profitable per order: Even with fewer orders, it gives the best money per order.

First Class underperforms: Slower and less profitable than Second Class, which is unusual for a premium service.

Same Day is a red flag: Slowest and least profitable—needs investigation or fixing.

### % of Shipments Exceeding Threshold

```python
threshold = 1500

Delay_Frequency = (
    (df['Shipping Lead Time'] > threshold).sum() / len(df)
) * 100

Delay_Frequency
```

```python
delay = (df['Shipping Lead Time'] > threshold).sum()
on_time = (df['Shipping Lead Time'] <= threshold).sum()

sizes = [delay, on_time]
labels = ['Exceeding Threshold', 'Within Threshold']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=60, pctdistance=0.85)

# Create donut hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

plt.title("Percentage of Shipments Exceeding Lead Time Threshold")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(
    x=df['Shipping Lead Time'] > threshold,
    color='purple',)

plt.xticks([0,1], ['Within Threshold', 'Exceeding Threshold'])
plt.title("Shipment Delay Distribution")
plt.ylabel("Number of Shipments")
plt.tight_layout()
plt.show()
```
